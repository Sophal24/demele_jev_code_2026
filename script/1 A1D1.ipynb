## Calculate the completeness of the survey

```python
result_dict = {}
for id in d1_df['A1']:
    try:
        tmp_df = a1_df[a1_df['A1'] == id]
        # print(tmp_df.shape[0])
        col_list = tmp_df.columns[3:]

        condition_dict = {}

        for col in col_list:
            # print(col,"=" , tmp_df[col].sum())
            condition_dict[col] = int(tmp_df[col].sum())
        condition_dict['Record in A1'] = tmp_df.shape[0]


        if (condition_dict['A4/1'] == 1 and condition_dict['A4/2'] == 1 
            and condition_dict['A4/3'] == 1 and condition_dict['A4/4'] == 1 
            and condition_dict['A4/5'] == 1 and condition_dict['A4/6'] == 1 
            and condition_dict['A4/7'] == 1 and condition_dict['A4/8'] == 1 
            and condition_dict['A4/9'] == 1):
            pass
        else:
            print(tmp_df)
            print(id)
            print(condition_dict)
            print(id, "is not completed! or ERROR")
            print()
            result_dict[id] = condition_dict

    except:
        print('NO ID')
    # break

print(result_dict)
```

```python
data = result_dict   # your JSON dictionary
print(data)

df = pd.DataFrame.from_dict(data, orient='index').reset_index()
df = df.rename(columns={'index':'ID'})
df['Remark'] = ""
df
```

```python
A4_cols = [f'A4/{i}' for i in range(1,10)]
df['A4_max'] = df[A4_cols].max(axis=1)
```

## Set up Remark Column

```python
df['Remark'] = np.select(
    [
        df['A4_max'] == 0,
        df['A4_max'] == 2,
        df['A4_max'] == 3,
        df['A4_max'] > 3
    ],
    [
        "All Missing",
        "Double Entry",
        "Triple Entry",
        "More Than 3 Entry"
    ],
    default=""
)

df
```

```python
df.replace({0: 'Missing'}, inplace=True)
df.drop(columns=['A4_max'], inplace=True) # drop column A4_max
df
```
