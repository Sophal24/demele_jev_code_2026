```python
def get_data_url(form_id):

    url = "https://demele-jev.uhs.edu.kh/api/download-submissions-v3"

    params = {
        "formId": form_id,
        "secret_key": "13ac3c2f-b7df-4ce9-8010-247da09a91d9",
        "splitSelectMultiples": "true",
        # "groupPaths": "false",
        "deletedFields": "true"
    }

    response = requests.get(
        url,
        params=params,
        timeout=10
    )

    response.raise_for_status()

    df = pd.read_csv(StringIO(response.text))

    return df
```
