```python
import requests
import pandas as pd
from requests.auth import HTTPBasicAuth
```

```python
# https://odkcentral2025.uhs.edu.kh/v1/projects/1/forms/A1_Questionnaire_Cohort1_Baseline.svc
```

```python
# ====== CONFIG ======
# BASE_URL = "https://your-domain.com" # https://odkcentral2025.uhs.edu.kh
BASE_URL = "https://odkcentral2025.uhs.edu.kh"
PROJECT_ID = 1
FORM_ID = "A1_Questionnaire_Cohort1_Baseline"

USERNAME = "sophal_thear@uhs.edu.kh"
PASSWORD = "uhs@123456"
```

```python
# ====== API URL ======
url = f"{BASE_URL}/v1/projects/{PROJECT_ID}/forms/{FORM_ID}/submissions"
```

```python
# ====== REQUEST ======
response = requests.get(
    url,
    auth=HTTPBasicAuth(USERNAME, PASSWORD),
    headers={"Accept": "application/json"}
)
```

```python
# ====== CHECK RESPONSE ======
if response.status_code == 200:
    data = response.json()
    df = pd.json_normalize(data)
    # print(df.head())
else:
    print("Error:", response.status_code, response.text)
```
