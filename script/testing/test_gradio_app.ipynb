```python
import gradio as gr
import pandas as pd

main_df = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

detail_data = {
    1: pd.DataFrame({"Feature": ["x", "y"], "Value": [10, 20]}),
    2: pd.DataFrame({"Feature": ["x", "y"], "Value": [300, 400]}),
    3: pd.DataFrame({"Feature": ["x", "y"], "Value": [5000, 6000]})
}

def show_details(evt: gr.SelectData):
    row_idx = evt.index[0]
    record_id = main_df.iloc[row_idx]["ID"]
    return detail_data[record_id]

with gr.Blocks() as demo:
    table = gr.DataFrame(main_df)
    details = gr.DataFrame(label="Details")

    table.select(show_details, outputs=details)

demo.launch()
```
