## Find duplicates rows and its error

```python
def find_duplicate_differences(df, id_col="A1"):

    result = []

    # only duplicated IDs
    duplicate_ids = df[df[id_col].duplicated(keep=False)][id_col].unique()

    for patient_id in duplicate_ids:

        tmp = df[df[id_col] == patient_id]

        diff_cols = []

        for col in tmp.columns:

            if col == id_col:
                continue

            # count unique values
            if tmp[col].nunique(dropna=False) > 1:
                diff_cols.append(col)

        result.append({
            id_col: patient_id,
            "Different Columns": diff_cols,
            "Number of Differences": len(diff_cols)
        })

    return pd.DataFrame(result)
```

```python
difference_report = find_duplicate_differences(merged_df)
difference_report
```
